**FILTRO 6**

En este sexto filtro decidi tomar la columna de "SUBTOTAL_PARTIDA" y filtrar en la columna aquellos totales que sean mayores o iguales a 50,000 pesos.

```python
#Filtro por comparacion
filtro6=df[df["SUBTOTAL_PARTIDA"] >= 50000]
filtro6
```

```python
filtro6.to_csv("Filtro6.csv")
```

**FILTRO 7**

En este septimo filtro decidi tomar de igual forma la columna de "SUBTOTAL_PARTIDA" y filtrar en la columna aquellos totales que sean mayores o iguales a 50,000 pesos, pero esta vez incluyendo la fecha en la que se creo o se genero el documento.

```python
#Filtro Y
filtro7=df[(df["SUBTOTAL_PARTIDA"] >= 50000) & (df["FECHA_DOC"] == "2022-09-23")]
filtro7
```

```python
filtro7.to_csv("Filtro7.csv")
```

**FILTRO 8**

En este octavo filtro decidi tomar de igual forma la columna de "SUBTOTAL_PARTIDA" y filtrar en la columna aquellos totales que sean mayores o iguales a 50,000 pesos. Pero esta vez filtra aquellos documentos en que su fecha de creación haya sido "2022-09-23" no importa su monto y aquellos documentos que en la columna de "SUBTOTAL_PARTIDA" su total haya sido mayor o igual a 50,000 pesos no importa su fecha. 

```python
#Filtro o (una u otra)
filtro8=df[(df["SUBTOTAL_PARTIDA"] >= 50000) | (df["FECHA_DOC"] == "2022-09-23")]
filtro8
```

```python
filtro8.to_csv("Filtro8.csv")
```

**FILTRO 9**

En este noveno filtro decidi tomar de igual forma la columna de "SUBTOTAL_PARTIDA" y filtrar en la columna aquellos totales que sean mayores o iguales a 50,000 pesos, pero esta vez filtra aquellos documentos que son mayores o iguales a 50,000 pesos y que no esten en la fecha de "2022-09-23" .

```python
#Filtro not
filtro9=df[~(df["SUBTOTAL_PARTIDA"] <= 50000) & ~(df["FECHA_DOC"] == "2022-09-23")]
filtro9
```

```python
filtro9.to_csv("Filtro9.csv")
```
